## Skenario 1: Filter Transaksi yang Lunas

````python
def is_lunas(transaksi):
    return transaksi['status'] == 'Lunas'

def filter_lunas(data):
    return list(filter(is_lunas, data))

transaksi_lunas = filter_lunas(transaksi)

print("Transaksi yang lunas:")

for t in transaksi_lunas:
    print(t)
    
# print(transaksi_lunas)
````

Dengan Lambda :

````python
print("Transaksi Lunas:"); 
print(*filter(lambda t: t['status'] == 'Lunas', transaksi), sep="\n")
````

## Skenario 2: Tambahkan Total Belanja dengan Map

````python
from typing import Any


def hitung_total(transaksi):
    return {**transaksi, 'total_belanja': transaksi['jumlah_item'] * transaksi['harga_per_item']}

def tambah_total_belanja(data): 
    return list[dict](map(hitung_total, data))

transaksi_lunas_dengan_total = tambah_total_belanja(transaksi_lunas)

for t in transaksi_lunas_dengan_total:
    print(t)
````

Dengan Lambda :

````python
print(*map(lambda t: {**t, 'total_belanja': t['jumlah_item'] * t['harga_per_item']},
           filter(lambda t: t['status'] == 'Lunas', transaksi)), sep="\n")
````

## Skenario 3: Hitung Total Pendapatan dengan Reduce

````python
from functools import reduce

def jumlahkan_total(accumulator, transaksi):
    return accumulator + transaksi['total_belanja']

def hitung_total_pendapatan(data):
    return reduce(jumlahkan_total, data, 0)

total_pendapatan = hitung_total_pendapatan(transaksi_lunas_dengan_total)

print(total_pendapatan)
````

Dengan Lambda :

````python
from functools import reduce
print(reduce(lambda acc, t: acc + t['jumlah_item'] * t['harga_per_item'],
             filter(lambda t: t['status'] == 'Lunas', transaksi), 0))
````

## Skenario 4: Transaksi Tidak Dibatalkan dan Hitung Total Item (Rekursif)

````python
def filter_tidak_dibatalkan(data):
    return [t for t in data if t['status'] != 'Batal']

def hitung_total_item_rekursif(data, index=0):
    if index >= len(data):
        return 0
    
    return data[index]['jumlah_item'] + hitung_total_item_rekursif(data, index + 1)

transaksi_tidak_dibatalkan = filter_tidak_dibatalkan(transaksi)

print("Transaksi yang tidak dibatalkan:")
for t in transaksi_tidak_dibatalkan:
    print(t)

print(f"\nTotal jumlah item dari transaksi tidak dibatalkan: {hitung_total_item_rekursif(transaksi_tidak_dibatalkan)}")
````

Dengan Lambda :

````python
print("Transaksi yang tidak dibatalkan:"); 
print(*filter(lambda t: t['status'] != 'Batal', transaksi), sep="\n")
````

2.   Buat sebuah factory function bernama buat_pencari_keyword(keyword). Fungsi ini harus mengembalikan sebuah fungsi closure baru yang menerima satu argumen teks. Fungsi closure ini akan "mengingat" keyword yang diberikan dan mengembalikan True jika teks (setelah diubah ke lowercase) mengandung keyword tersebut, dan False jika tidak. Buat dua instance: cari_bagus = buat_pencari_keyword('bagus') dan cari_jelek = buat_pencari_keyword('jelek').

Contoh Output:



```
sentimen positif =  ['produk ini sangat bagus! Kualitasnya mantap.', 'Bagus sih, tapi pengirimannya lama.']
sentimen negatif =  ['jelek sekali, jangan beli. sangat mengecewakan.']
```



````python
def buat_pencari_keyword(keyword):
    def pencari(teks):
        return keyword.lower() in teks.lower()
    return pencari

cari_bagus = buat_pencari_keyword('bagus')
cari_jelek = buat_pencari_keyword('jelek')

sentimen_positif = [ulasan['teks'].lower() for ulasan in data_ulasan if cari_bagus(ulasan['teks'])]
sentimen_negatif = [ulasan['teks'].lower() for ulasan in data_ulasan if cari_jelek(ulasan['teks'])]

print("sentimen positif =", sentimen_positif)
print("sentimen negatif =", sentimen_negatif)
````





3. Buat sebuah Decorator bernama @sanitasi_teks. Decorator ini harus secara otomatis membersihkan argumen string pertama yang diterima oleh fungsi yang dihiasnya (mengubahnya ke lowercase dan .strip()) sebelum meneruskannya ke fungsi asli. Terapkan decorator ini pada fungsi hitung_jumlah_kata(teks) di bawah ini.



````python

def sanitasi_teks(func):
    def wrapper(teks, *args, **kwargs):
        teks_bersih = teks.lower().strip()
        return func(teks_bersih, *args, **kwargs)
    return wrapper

@sanitasi_teks
def hitung_jumlah_kata(teks):
    print(f"DEBUG: Menghitung kata dari: '{teks}'")
    return len(teks.split())
````

````python
# Panggil fungsi untuk setiap ulasan
for ulasan in data_ulasan:
    jumlah = hitung_jumlah_kata(ulasan['teks'])
    print(f"jumlah kata= {jumlah}")
````

````python
for ulasan in data_ulasan:
    teks = ulasan['teks']
    jumlah_kata = hitung_jumlah_kata(teks)
    print(f"jumlah kata= {jumlah_kata}")
````
