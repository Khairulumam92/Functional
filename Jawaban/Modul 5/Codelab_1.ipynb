## Skenario 1: Filter Transaksi yang Lunas

```python
# Pure function untuk mengecek status lunas
def is_lunas(transaksi):
    """Mengecek apakah transaksi berstatus Lunas"""
    return transaksi['status'] == 'Lunas'

# Pure function untuk filter transaksi lunas
def filter_lunas(data):
    """Filter transaksi dengan status 'Lunas' menggunakan filter"""
    return list(filter(is_lunas, data))

# Skenario 1: Filter transaksi lunas
transaksi_lunas = filter_lunas(transaksi)

# print("Transaksi yang lunas:")
# for t in transaksi_lunas:
#     print(t)
    
print(transaksi_lunas)
```

## Skenario 2: Tambahkan Total Belanja dengan Map

```python
# Pure function untuk menghitung total belanja satu transaksi
from typing import Any


def hitung_total(transaksi):
    """Menghitung total belanja dari satu transaksi"""
    return {**transaksi, 'total_belanja': transaksi['jumlah_item'] * transaksi['harga_per_item']}

# Pure function untuk menambahkan total belanja
def tambah_total_belanja(data):
    """Menambahkan field total_belanja menggunakan map"""
    return list[dict](map(hitung_total, data))

# Skenario 2: Tambahkan total belanja pada transaksi lunas
transaksi_lunas_dengan_total = tambah_total_belanja(transaksi_lunas)

for t in transaksi_lunas_dengan_total:
    print(t)
```

## Skenario 3: Hitung Total Pendapatan dengan Reduce

```python
from functools import reduce

# Pure function untuk menjumlahkan total belanja
def jumlahkan_total(accumulator, transaksi):
    """Menjumlahkan total belanja dari transaksi ke accumulator"""
    return accumulator + transaksi['total_belanja']

# Pure function untuk menghitung total pendapatan
def hitung_total_pendapatan(data):
    """Menghitung total pendapatan dari field total_belanja menggunakan reduce"""
    return reduce(jumlahkan_total, data, 0)

# Skenario 3: Hitung total pendapatan
total_pendapatan = hitung_total_pendapatan(transaksi_lunas_dengan_total)

print(total_pendapatan)
```

## Skenario 4: Transaksi Tidak Dibatalkan dan Hitung Total Item (Rekursif)

```python
# Pure function untuk filter transaksi tidak dibatalkan menggunakan list comprehension
def filter_tidak_dibatalkan(data):
    """Filter transaksi yang tidak dibatalkan (status != 'Batal')"""
    return [t for t in data if t['status'] != 'Batal']

# Pure function rekursif untuk menghitung total jumlah item
def hitung_total_item_rekursif(data, index=0):
    """Menghitung total jumlah_item secara rekursif"""
    # Base case: jika sudah mencapai akhir list
    if index >= len(data):
        return 0
    
    # Recursive case: tambahkan jumlah_item pada index saat ini dengan hasil rekursi selanjutnya
    return data[index]['jumlah_item'] + hitung_total_item_rekursif(data, index + 1)

# Skenario 4: Filter transaksi tidak dibatalkan
transaksi_tidak_dibatalkan = filter_tidak_dibatalkan(transaksi)

print("Transaksi yang tidak dibatalkan:")
for t in transaksi_tidak_dibatalkan:
    print(t)

print(f"\nTotal jumlah item dari transaksi tidak dibatalkan: {hitung_total_item_rekursif(transaksi_tidak_dibatalkan)}")
```

## Visualisasi 2: Pie Chart - Persentase Status Transaksi


```python
# Membuat Pie Chart yang menarik dengan explode, label, dan nilai
fig, ax = plt.subplots(figsize=(10, 8))

# Warna yang menarik untuk setiap kategori
colors_pie = ['#2ecc71', '#f39c12', '#e74c3c']  # Hijau untuk Lunas, Oranye untuk Belum Lunas, Merah untuk Batal

# Explode untuk menarik perhatian pada slice terbesar (menggunakan lambda)
# Explode slice dengan persentase terbesar
max_index = status_percentages.index(max(status_percentages))
explode_values = list(map(lambda i: 0.1 if i == max_index else 0, range(len(status_labels))))

# Membuat autopct dengan format yang menarik (menggunakan lambda)
autopct_func = lambda pct: f'{pct:.1f}%\n({int(pct/100 * total_transaksi)} transaksi)'

# Membuat pie chart
wedges, texts, autotexts = ax.pie(
    status_counts,
    labels=status_labels,
    colors=colors_pie,
    explode=explode_values,
    autopct=autopct_func,
    shadow=True,
    startangle=90,
    textprops={'fontsize': 11, 'fontweight': 'bold'},
    pctdistance=0.85,
    labeldistance=1.05
)

# Styling untuk autotext (nilai persentase)
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')

# Menambahkan lingkaran di tengah untuk donut chart effect
centre_circle = plt.Circle((0, 0), 0.70, fc='white', edgecolor='gray', linewidth=2)
ax.add_artist(centre_circle)

# Title
ax.set_title('Persentase Status\nDistribusi Status Transaksi Supermarket', 
             fontsize=14, fontweight='bold', pad=20)

# Menambahkan informasi tambahan di bawah
info_text = f'Total Transaksi: {total_transaksi}'
ax.text(0, -1.3, info_text, ha='center', fontsize=10, style='italic',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()
plt.show()

print(f"\nPie Chart menunjukkan distribusi status dari {total_transaksi} transaksi.")
print("Slice dengan explode menunjukkan status yang paling dominan.")
```

```python
# Membuat visualisasi kombinasi seperti pada contoh gambar
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# ========== Scatter Plot (Kiri) ==========
max_belanja = max(y_data)
sizes_scatter = [(belanja / max_belanja) * 300 + 50 for belanja in y_data]
colors_scatter = list(map(lambda y: 'darkblue' if y >= 150000 else ('steelblue' if y >= 120000 else 'lightblue'), y_data))

ax1.scatter(x_data, y_data, s=sizes_scatter, c=colors_scatter, alpha=0.7, 
            edgecolors='black', linewidth=1.5)

ax1.set_xlabel('Jumlah Item', fontsize=11, fontweight='bold')
ax1.set_ylabel('Total Belanja (Rp)', fontsize=11, fontweight='bold')
ax1.set_title('Ini Scatter Plot', fontsize=13, fontweight='bold', pad=15)
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):,}'))
ax1.grid(True, alpha=0.3, linestyle='--')
ax1.set_facecolor('#f8f9fa')

# ========== Pie Chart (Kanan) ==========
# Mapping status untuk label yang lebih sesuai dengan contoh
status_labels_pie = ['Berhasil', 'Pending', 'Gagal']  # Berhasil = Lunas, Pending = Belum Lunas, Gagal = Batal
status_counts_pie = [status_lunas, status_belum_lunas, status_batal]
colors_pie_combined = ['#3498db', '#f39c12', '#27ae60']  # Biru, Oranye, Hijau

max_index_pie = status_percentages.index(max(status_percentages))
explode_combined = list(map(lambda i: 0.1 if i == max_index_pie else 0, range(len(status_labels_pie))))

autopct_combined = lambda pct: f'{pct:.1f}%'

wedges, texts, autotexts = ax2.pie(
    status_counts_pie,
    labels=status_labels_pie,
    colors=colors_pie_combined,
    explode=explode_combined,
    autopct=autopct_combined,
    shadow=True,
    startangle=90,
    textprops={'fontsize': 11, 'fontweight': 'bold'},
    pctdistance=0.85
)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')

ax2.set_title('Persentase Status', fontsize=13, fontweight='bold', pad=15)

plt.suptitle('Visualisasi Data Transaksi Supermarket', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("Visualisasi kombinasi selesai!")
```
