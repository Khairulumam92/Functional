Telah disediakan sebuah data yang perlu diolah dengan menggunakan konsep lazy-functional. Anda diminta untuk menyelesaikan ketiga masalah di bawah ini dengan memanfaatkan generator.
Ketentuan Pengerjaan:
- Anda wajib menggunakan kedua jenis implementasi generator: fungsi generator (fungsi yang menggunakan yield) dan generator expression.
- Dari tiga soal yang diberikan, minimal harus ada satu soal yang diselesaikan menggunakan fungsi generator dan minimal satu soal yang diselesaikan menggunakan generator expression.
- Anda dibebaskan untuk memilih soal mana yang akan dikerjakan dengan fungsi generator ataupun generator expression, selama kedua ketentuan di atas terpenuhi. (Contoh: 1 fungsi generator & 2 generator expression, atau 2 fungsi generator & 1 generator expression).
- Untuk menampilkan hasil, Anda wajib mendemonstrasikan cara kerja generator secara manual. Gunakan perulangan while dan next() untuk mengambil setiap nilai, serta terapkan try-except StopIteration untuk menangani kondisi ketika generator telah habis dan menghentikan perulangan, dan tambahkan print untuk mengetahui apakah iterator sudah selesai. Dilarang menggunakan perulangan for untuk menampilkan hasil akhir.

1. Buatlah generator generator/generator expression yang hanya menghasilkan data
karyawan dengan status_aktif == True.

```python
# Moh. Khairul Umam - 202310370311448
# TUGAS LAZY-FUNCTIONAL DENGAN GENERATOR

# Data karyawan
karyawan = [
    {'nama': 'Zaky', 'gaji': 5000000, 'bonus': 1000000, 'status_aktif': True},
    {'nama': 'Fitra', 'gaji': 4500000, 'bonus': None, 'status_aktif': True},
    {'nama': 'Fia', 'gaji': 5200000, 'bonus': 500000, 'status_aktif': True},
    {'nama': 'Adit', 'gaji': 6000000, 'bonus': 1000000, 'status_aktif': False},
    {'nama': 'Faizal', 'gaji': 4000000, 'bonus': 700000, 'status_aktif': True},
    {'nama': 'Radan', 'gaji': 5500000, 'bonus': 'tidak ada', 'status_aktif': True},
    {'nama': 'Wempy', 'gaji': 4800000, 'bonus': 600000, 'status_aktif': True},
    {'nama': 'Alfi', 'gaji': 5000000, 'bonus': 800000, 'status_aktif': False},
    {'nama': 'Hakim', 'gaji': 5200000, 'bonus': 1000000, 'status_aktif': True},
    {'nama': 'Rama', 'gaji': None, 'bonus': 750000, 'status_aktif': True}
]

print("="*60)
print("SOAL 1: Filter Karyawan Aktif (Generator Expression)")
print("="*60)

# Generator expression untuk filter karyawan dengan status_aktif == True
karyawan_aktif_gen = (k for k in karyawan if k['status_aktif'] == True)

# Menampilkan hasil menggunakan while + next() + try-except
hasil_karyawan_aktif = []
while True:
    try:
        data = next(karyawan_aktif_gen)
        hasil_karyawan_aktif.append(data)
        print(data)
    except StopIteration:
        print("\n[SELESAI] Iterator selesai - Semua karyawan aktif telah ditampilkan")
        break

print(f"\nTotal karyawan aktif: {len(hasil_karyawan_aktif)}")
```

2. Buatlah sebuah generator/generator expression yang menghasilkan nama karyawan dengan data yang tidak valid (gaji atau bonus bukan integer).

```python
# Moh. Khairul Umam - 202310370311448
print("\n" + "="*60)
print("SOAL 2: Nama Karyawan dengan Data Tidak Valid (Fungsi Generator)")
print("="*60)

# Fungsi generator untuk mendapatkan nama karyawan dengan data tidak valid
def generator_karyawan_invalid(data_karyawan):
    """
    Generator yang menghasilkan nama karyawan dengan data tidak valid
    (gaji atau bonus bukan integer)
    """
    for k in data_karyawan:
        # Cek apakah gaji bukan integer atau None
        gaji_invalid = not isinstance(k['gaji'], int) or k['gaji'] is None
        # Cek apakah bonus bukan integer (None atau tipe lain)
        bonus_invalid = not isinstance(k['bonus'], int) or k['bonus'] is None
        
        if gaji_invalid or bonus_invalid:
            yield k['nama']

# Membuat instance generator
nama_invalid_gen = generator_karyawan_invalid(karyawan)

# Menampilkan hasil menggunakan while + next() + try-except
while True:
    try:
        nama = next(nama_invalid_gen)
        print(nama)
    except StopIteration:
        print("\n[SELESAI] Iterator selesai - Semua nama karyawan dengan data tidak valid telah ditampilkan")
        break
```

3. Buatlah sebuah generator yang menghitung kontribusi tiap karyawan aktif dengan
data valid terhadap total kompensasi (gaji + bonus). Hasil generator harus berupa
dictionary dengan key:
- a."nama" sebagai nama karyawan
- b."kompensasi" sebagai total gaji ditambah bonus
- c."kontribusi" sebagai persentase kontribusi terhadap total kompensasi semua karyawan valid (format string xx.xx%)

```python
print("\n" + "="*60)
print("SOAL 3: Kontribusi Karyawan Aktif (Fungsi Generator)")
print("="*60)

# Fungsi generator untuk menghitung kontribusi karyawan
def generator_kontribusi_karyawan(data_karyawan):
    """
    Generator yang menghasilkan dictionary kontribusi tiap karyawan aktif
    dengan data valid terhadap total kompensasi
    """
    # Pertama, filter karyawan aktif dengan data valid dan hitung total kompensasi
    karyawan_valid = []
    total_kompensasi_all = 0
    
    for k in data_karyawan:
        # Hanya proses karyawan yang aktif
        if k['status_aktif']:
            # Validasi data: gaji dan bonus harus integer
            if isinstance(k['gaji'], int) and isinstance(k['bonus'], int):
                kompensasi = k['gaji'] + k['bonus']
                karyawan_valid.append({
                    'nama': k['nama'],
                    'kompensasi': kompensasi
                })
                total_kompensasi_all += kompensasi
    
    # Yield hasil dengan perhitungan kontribusi
    for k in karyawan_valid:
        kontribusi_persen = (k['kompensasi'] / total_kompensasi_all) * 100
        yield {
            'nama': k['nama'],
            'kompensasi': k['kompensasi'],
            'kontribusi': f"{kontribusi_persen:.2f}%"
        }

# Membuat instance generator
kontribusi_gen = generator_kontribusi_karyawan(karyawan)

# Menampilkan hasil menggunakan while + next() + try-except
hasil_kontribusi = []
while True:
    try:
        data = next(kontribusi_gen)
        hasil_kontribusi.append(data)
        print(data)
    except StopIteration:
        print("\n[SELESAI] Iterator selesai - Semua kontribusi karyawan telah ditampilkan")
        break

print(f"\nTotal karyawan valid yang dihitung: {len(hasil_kontribusi)}")
```
