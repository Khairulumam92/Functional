# SOAL 6: Modul 4 - Closure dan Factory Function

**Soal:**
Buatlah factory function untuk membuat validator dengan closure:

1. `create_range_validator(min_val, max_val)` - Mengembalikan fungsi validator yang mengecek apakah nilai berada dalam range
2. `create_length_validator(min_len, max_len)` - Mengembalikan fungsi validator untuk panjang string
3. `create_composite_validator(*validators)` - Mengembalikan validator yang menjalankan semua validator dan mengembalikan True jika semua valid


**Contoh:**
```python
age_validator = create_range_validator(18, 65)
print(age_validator(25))  # True
print(age_validator(70))   # False

name_validator = create_length_validator(3, 20)
print(name_validator("John"))  # True
print(name_validator("Jo"))    # False

composite = create_composite_validator(age_validator, name_validator)
print(composite(25, "John"))  # True
```

**Petunjuk:**
- Gunakan closure untuk menyimpan min_val dan max_val
- Composite validator harus menjalankan semua validator yang diberikan

# SOAL 8: Modul 4 - Decorator

**Soal:**
Buatlah decorator untuk:

1. `@timer` - Mencetak waktu eksekusi fungsi
2. `@logger` - Mencetak nama fungsi dan argumen yang dipanggil
3. `@retry(n)` - Mengulang eksekusi fungsi maksimal n kali jika terjadi error
4. `@validate_input(validator_func)` - Validasi input sebelum eksekusi fungsi


**Contoh:**
```python
@timer
@logger
def add(a, b):
    return a + b

@retry(3)
def risky_operation():
    # ... operation that might fail
    pass

@validate_input(lambda x: x > 0)
def square(x):
    return x ** 2
```

**Petunjuk:**
- Decorator adalah fungsi yang menerima fungsi dan mengembalikan fungsi
- Decorator dengan parameter memerlukan nested functions
- Gunakan functools.wraps untuk mempertahankan metadata fungsi
