## BAGIAN 1: Pure Functions untuk Data Processing

### 1. Hitung Total Revenue per Item

```python
# Pure function untuk menghitung total revenue per item
def hitung_total_revenue(menu):
    return menu["harga"] * menu["jumlah"]

def rating_rata(rate):
    return rate["rating"] * rate["jumlah"] / rate["jumlah"] 

rata_rate = {"rating": 4.8, "jumlah": 150, "jumlah": 150}
rata = rating_rata(rata_rate)
print(f"total {rata:,}")

# Test
item_test = {"harga": 25000, "jumlah": 150}
total = hitung_total_revenue(item_test)
print(f"Total revenue untuk item test: Rp {total:,}")
```

### 2. Tambahkan Field Total Revenue dengan Map

```python
# Pure function untuk menambahkan field total_revenue menggunakan map
def tambah_total_revenue(data):
    """Menambahkan field total_revenue ke setiap item menggunakan map dan lambda"""
    return list(map(
        lambda item: {**item, "total_revenue": hitung_total_revenue(item)},
        data
    ))

# Test
data_dengan_revenue = tambah_total_revenue(data_penjualan)
print(f"Data pertama dengan total_revenue: {data_dengan_revenue[0]}")
print(f"\nData asli tidak berubah (tidak ada field total_revenue): {data_penjualan[0]}")
```

### 3. Filter Data Berdasarkan Kategori dengan Filter dan Lambda

```python
# Pure function untuk filter berdasarkan kategori menggunakan filter dan lambda
def filter_kategori(data, kategori):
    """Memfilter data berdasarkan kategori menggunakan filter dan lambda"""
    return list(filter(lambda item: item["kategori"] == kategori, data))

# Test
data_makanan = filter_kategori(data_penjualan, "Makanan")
data_minuman = filter_kategori(data_penjualan, "Minuman")
data_dessert = filter_kategori(data_penjualan, "Dessert")

print(f"Jumlah data Makanan: {len(data_makanan)}")
print(f"Jumlah data Minuman: {len(data_minuman)}")
print(f"Jumlah data Dessert: {len(data_dessert)}")
```

### 4. Hitung Total Omset Keseluruhan dengan Reduce

```python
# Pure function untuk menghitung total omset menggunakan reduce
def hitung_total_omset(data):
    """Menghitung total omset semua penjualan menggunakan reduce"""
    return reduce(
        lambda acc, item: acc + (item["harga"] * item["jumlah"]),
        data,
        0
    )

# Test
total_omset = hitung_total_omset(data_penjualan)
print(f"Total omset keseluruhan: Rp {total_omset:,}")
```

### 5. Kelompokkan Data per Bulan dengan List Comprehension

```python
# Pure function untuk mengelompokkan data per bulan menggunakan list comprehension
def kelompokkan_per_bulan(data):
    """Mengelompokkan data per bulan dan menghitung total omset per bulan"""
    # Ambil unique bulan
    bulan_list = list(set(item["bulan"] for item in data))
    
    # Hitung total omset per bulan menggunakan list comprehension dan reduce
    return {
        bulan: reduce(
            lambda acc, item: acc + (item["harga"] * item["jumlah"]),
            [item for item in data if item["bulan"] == bulan],
            0
        )
        for bulan in bulan_list
    }

# Test
omset_per_bulan = kelompokkan_per_bulan(data_penjualan)
print("Total omset per bulan:")
for bulan, omset in omset_per_bulan.items():
    print(f"  {bulan}: Rp {omset:,}")
```

### 6. Factory Function dengan Closure

```python
# Factory function dengan closure untuk membuat filter harga
def buat_filter_harga(min_harga, max_harga):
    """Factory function dengan closure untuk membuat filter berdasarkan range harga"""
    def filter_func(item):
        return min_harga <= item["harga"] <= max_harga
    return filter_func

# Test
filter_murah = buat_filter_harga(0, 15000)
filter_sedang = buat_filter_harga(15001, 20000)
filter_mahal = buat_filter_harga(20001, float('inf'))

data_murah = list(filter(filter_murah, data_penjualan))
data_sedang = list(filter(filter_sedang, data_penjualan))
data_mahal = list(filter(filter_mahal, data_penjualan))

print(f"Produk murah (‚â§ Rp 15.000): {len(data_murah)} item")
print(f"Produk sedang (Rp 15.001 - Rp 20.000): {len(data_sedang)} item")
print(f"Produk mahal (> Rp 20.000): {len(data_mahal)} item")
```

### 7. Generator Expression

```python
# Generator expression untuk menghasilkan total revenue per item
def generator_total_revenue(data):
    """Generator expression untuk menghasilkan total revenue per item"""
    return (item["harga"] * item["jumlah"] for item in data)

# Test - tampilkan 5 item pertama
gen = generator_total_revenue(data_penjualan)
print("5 Total revenue pertama (dari generator):")
for i, total in enumerate(gen):
    if i < 5:
        print(f"  Item {i+1}: Rp {total:,}")
    else:
        break

# Atau menggunakan list untuk menampilkan semua (tapi ini bukan generator lagi)
print("\nSemua total revenue (dari generator, dikonversi ke list):")
all_revenues = list(generator_total_revenue(data_penjualan))
print(f"Total items: {len(all_revenues)}")
print(f"Total omset (dari generator): Rp {sum(all_revenues):,}")
```

## BAGIAN 2: Analisis Data dengan List Comprehension

### 8. Rata-rata Rating per Kategori

```python
# Pure function untuk menghitung rata-rata rating per kategori
def rata_rata_rating_per_kategori(data):
    """Menghitung rata-rata rating per kategori menggunakan list comprehension dan reduce"""
    # Ambil unique kategori
    kategori_list = list(set(item["kategori"] for item in data))
    
    # Hitung rata-rata rating per kategori
    result = {}
    for kategori in kategori_list:
        # Filter data per kategori
        data_kategori = [item for item in data if item["kategori"] == kategori]
        
        # Hitung total rating dan jumlah item
        total_rating = reduce(lambda acc, item: acc + item["rating"], data_kategori, 0)
        jumlah_item = len(data_kategori)
        
        # Hitung rata-rata
        result[kategori] = total_rating / jumlah_item if jumlah_item > 0 else 0
    
    return result

# Test
rating_per_kategori = rata_rata_rating_per_kategori(data_penjualan)
print("Rata-rata rating per kategori:")
for kategori, rating in rating_per_kategori.items():
    print(f"  {kategori}: {rating:.2f}")
```

### 9. Menu Terlaris per Bulan

```python
# Pure function untuk menemukan menu terlaris per bulan
def menu_terlaris_per_bulan(data):
    """Menemukan menu dengan jumlah terjual terbanyak per bulan menggunakan list comprehension"""
    # Ambil unique bulan
    bulan_list = list(set(item["bulan"] for item in data))
    
    result = {}
    for bulan in bulan_list:
        # Filter data per bulan
        data_bulan = [item for item in data if item["bulan"] == bulan]
        
        # Cari menu dengan jumlah terjual terbanyak
        menu_terlaris = max(data_bulan, key=lambda item: item["jumlah"])
        result[bulan] = (menu_terlaris["menu"], menu_terlaris["jumlah"])
    
    return result

# Test
terlaris_per_bulan = menu_terlaris_per_bulan(data_penjualan)
print("Menu terlaris per bulan:")
for bulan, (menu, jumlah) in terlaris_per_bulan.items():
    print(f"  {bulan}: {menu} ({jumlah} pcs)")
```

## BAGIAN 3: Visualisasi Data dengan Matplotlib

```python
# Persiapan data untuk visualisasi
# Data dengan total_revenue
data_dengan_revenue = tambah_total_revenue(data_penjualan)

# 1. Data untuk Bar Chart: Total omset per kategori
kategori_list = list(set(item["kategori"] for item in data_penjualan))
omset_per_kategori = {
    kategori: reduce(
        lambda acc, item: acc + (item["harga"] * item["jumlah"]),
        filter_kategori(data_penjualan, kategori),
        0
    )
    for kategori in kategori_list
}

# 2. Data untuk Pie Chart: Distribusi penjualan per bulan
omset_per_bulan = kelompokkan_per_bulan(data_penjualan)
total_omset_all = hitung_total_omset(data_penjualan)
persentase_per_bulan = {
    bulan: (omset / total_omset_all) * 100
    for bulan, omset in omset_per_bulan.items()
}

# 3. Data untuk Line Chart: Trend omset per bulan
bulan_order = ["Januari", "Februari", "Maret", "April"]
omset_trend = [omset_per_bulan.get(bulan, 0) for bulan in bulan_order]

# 4. Data untuk Scatter Plot: Harga vs Jumlah terjual
harga_list = [item["harga"] for item in data_penjualan]
jumlah_list = [item["jumlah"] for item in data_penjualan]
rating_list = [item["rating"] for item in data_penjualan]
kategori_list_scatter = [item["kategori"] for item in data_penjualan]

print("Data untuk visualisasi sudah siap!")
```

```python
# Buat visualisasi dengan 4 subplot
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Analisis Data Penjualan Restoran', fontsize=16, fontweight='bold')

# Warna untuk kategori
warna_kategori = {
    "Makanan": '#FF6B6B',
    "Minuman": '#4ECDC4',
    "Dessert": '#45B7D1'
}

# Warna untuk bulan
warna_bulan = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#F7B731']

# ===== SUBPLOT 1: Bar Chart - Total Omset per Kategori =====
kategori_names = list(omset_per_kategori.keys())
omset_values = [omset_per_kategori[k] for k in kategori_names]
colors_bar = [warna_kategori[k] for k in kategori_names]

bars = axes[0, 0].bar(kategori_names, omset_values, color=colors_bar)
axes[0, 0].set_title('Total Omset per Kategori', fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('Kategori')
axes[0, 0].set_ylabel('Total Omset (Juta Rupiah)')
axes[0, 0].grid(axis='y', alpha=0.3)

# Format y-axis dengan pemisah ribuan
axes[0, 0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.1f}M'))

# Tambahkan label nilai di atas setiap bar
for bar in bars:
    height = bar.get_height()
    axes[0, 0].text(bar.get_x() + bar.get_width()/2., height,
                   f'Rp {height:,.0f}',
                   ha='center', va='bottom', fontsize=9)

# ===== SUBPLOT 2: Pie Chart - Distribusi Penjualan per Bulan =====
bulan_names = list(persentase_per_bulan.keys())
persentase_values = [persentase_per_bulan[b] for b in bulan_order]

axes[0, 1].pie(persentase_values, labels=bulan_order, autopct='%1.1f%%', 
               colors=warna_bulan, startangle=90)
axes[0, 1].set_title('Distribusi Penjualan per Bulan', fontsize=12, fontweight='bold')

# ===== SUBPLOT 3: Line Chart - Trend Omset per Bulan =====
axes[1, 0].plot(bulan_order, omset_trend, marker='o', linewidth=2, 
                color='#FF6B6B', markersize=8)
axes[1, 0].set_title('Trend Omset per Bulan', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('Bulan')
axes[1, 0].set_ylabel('Total Omset (Rupiah)')
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'Rp {x:,.0f}'))

# Rotate x-axis labels
axes[1, 0].tick_params(axis='x', rotation=45)

# ===== SUBPLOT 4: Scatter Plot - Harga vs Jumlah Terjual =====
kategori_unique = list(set(kategori_list_scatter))
for kategori in kategori_unique:
    # Filter data per kategori
    indices = [i for i, k in enumerate(kategori_list_scatter) if k == kategori]
    harga_kategori = [harga_list[i] for i in indices]
    jumlah_kategori = [jumlah_list[i] for i in indices]
    rating_kategori = [rating_list[i] for i in indices]
    
    # Size berdasarkan rating
    sizes = [r * 50 for r in rating_kategori]
    
    axes[1, 1].scatter(harga_kategori, jumlah_kategori, 
                     s=sizes, alpha=0.6, label=kategori,
                     color=warna_kategori[kategori])

axes[1, 1].set_title('Hubungan Harga vs Jumlah Terjual', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('Harga Produk (Rupiah)')
axes[1, 1].set_ylabel('Jumlah Terjual')
axes[1, 1].legend()
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'Rp {x:,.0f}'))

# Adjust layout
plt.tight_layout()
plt.show()

print("\nVisualisasi selesai!")
```

**Teknik Functional Programming yang Digunakan:**

1. ‚úÖ **Pure Functions**: Semua fungsi tidak memiliki side effects
2. ‚úÖ **Map**: Untuk transformasi data (menambahkan field total_revenue)
3. ‚úÖ **Filter**: Untuk seleksi data berdasarkan kategori
4. ‚úÖ **Reduce**: Untuk agregasi data (total omset, rata-rata)
5. ‚úÖ **Lambda**: Digunakan dalam map, filter, dan reduce
6. ‚úÖ **List Comprehension**: Untuk filtering dan transformasi data
7. ‚úÖ **Closure**: Factory function untuk membuat filter harga
8. ‚úÖ **Generator Expression**: Untuk menghasilkan total revenue tanpa menyimpan semua data
9. ‚úÖ **Immutability**: Semua update menggunakan immutable approach (dictionary baru)
10. ‚úÖ **Visualisasi**: Menggunakan matplotlib dengan data yang sudah diproses secara functional
