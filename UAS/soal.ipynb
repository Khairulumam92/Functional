# SOAL 1: Sistem Analisis Nilai Mahasiswa

**Deskripsi:**
Anda diminta membuat sistem analisis nilai mahasiswa dengan pendekatan functional programming. Sistem ini harus dapat memproses data nilai dan menghasilkan berbagai statistik.

**Data yang diberikan:**

```python
mahasiswa = [
    {"nim": "001", "nama": "Andi", "nilai": [85, 90, 78, 88]},
    {"nim": "002", "nama": "Budi", "nilai": [70, 65, 72, 68]},
    {"nim": "003", "nama": "Citra", "nilai": [95, 92, 98, 96]},
    {"nim": "004", "nama": "Dina", "nilai": [60, 55, 58, 62]},
    {"nim": "005", "nama": "Eko", "nilai": [80, 85, 82, 78]}
]
```

**Tugas:**

### A. Pure Functions (20 poin)
Buatlah pure functions berikut:
1. `hitung_rata_rata(nilai_list)` - Menghitung rata-rata dari list nilai
2. `tentukan_grade(rata_rata)` - Menentukan grade berdasarkan rata-rata:
   - A: >= 85
   - B: >= 70
   - C: >= 55
   - D: < 55

### B. Higher-Order Functions dengan Lambda (25 poin)
Gunakan **map**, **filter**, dan **lambda** untuk:
1. Menambahkan field `rata_rata` dan `grade` ke setiap mahasiswa (gunakan **map** dengan **lambda**)
2. Memfilter mahasiswa yang mendapat grade A atau B (gunakan **filter** dengan **lambda**)
3. Mengekstrak hanya nama mahasiswa yang lulus (grade bukan D) menggunakan kombinasi **filter** dan **map**

### C. Closure - Factory Function (25 poin)
Buatlah closure/factory function:
1. `create_grade_filter(min_grade)` - Mengembalikan fungsi yang memfilter mahasiswa berdasarkan grade minimum
   - Contoh: `filter_grade_B = create_grade_filter('B')` akan memfilter mahasiswa dengan grade B ke atas

2. `create_nilai_adjuster(adjustment)` - Mengembalikan fungsi yang menambahkan nilai adjustment ke semua nilai mahasiswa
   - Contoh: `tambah_5 = create_nilai_adjuster(5)` akan menambah 5 ke setiap nilai

### D. Sequence Type Operations (15 poin)
Gunakan **list comprehension** untuk:
1. Membuat list tuple `(nama, rata_rata)` untuk semua mahasiswa
2. Membuat dictionary `{nim: grade}` untuk semua mahasiswa

### E. Reduce untuk Agregasi (15 poin)
Gunakan **reduce** untuk:
1. Menghitung total semua nilai rata-rata mahasiswa
2. Mencari mahasiswa dengan rata-rata tertinggi

---

**Contoh Output yang Diharapkan:**

```
=== Data Mahasiswa dengan Grade ===
Andi - Rata-rata: 85.25, Grade: A
Budi - Rata-rata: 68.75, Grade: C
...

=== Mahasiswa Grade A/B ===
['Andi', 'Citra', 'Eko']

=== Filter dengan Closure (Grade B ke atas) ===
['Andi', 'Citra', 'Eko']

=== Nilai setelah adjustment +5 ===
Andi: [90, 95, 83, 93]
...

=== List Tuple (nama, rata_rata) ===
[('Andi', 85.25), ('Budi', 68.75), ...]

=== Dictionary {nim: grade} ===
{'001': 'A', '002': 'C', ...}

=== Total Rata-rata: 377.25 ===
=== Mahasiswa Terbaik: Citra dengan rata-rata 95.25 ===
```

# SOAL 2: Sistem Manajemen Toko Online

**Deskripsi:**
Anda diminta membuat sistem manajemen toko online sederhana dengan pendekatan functional programming. Sistem ini harus dapat memproses data produk, menghitung diskon, dan menghasilkan laporan penjualan.

**Data yang diberikan:**

```python
produk = [
    {"id": "P001", "nama": "Laptop Gaming", "harga": 15000000, "stok": 10, "kategori": "Elektronik"},

    {"id": "P005", "nama": "Headset Gaming", "harga": 1200000, "stok": 25, "kategori": "Aksesoris"},
    {"id": "P006", "nama": "Webcam HD", "harga": 500000, "stok": 0, "kategori": "Elektronik"}
]

transaksi = [
    {"id_produk": "P001", "jumlah": 2},
    {"id_produk": "P002", "jumlah": 5},
    {"id_produk": "P003", "jumlah": 3},
    {"id_produk": "P005", "jumlah": 4}
]
```

**Tugas:**

### A. Pure Functions (20 poin)
Buatlah pure functions berikut:
1. `hitung_diskon(harga, persen_diskon)` - Menghitung harga setelah diskon
2. `hitung_total_nilai_stok(produk_item)` - Menghitung total nilai stok (harga x stok) untuk satu produk
3. `cek_ketersediaan(produk_item, jumlah)` - Mengecek apakah stok mencukupi (return True/False)

### B. Higher-Order Functions dengan Lambda (25 poin)
Gunakan **map**, **filter**, dan **lambda** untuk:
1. Menambahkan field `nilai_stok` (harga x stok) ke setiap produk menggunakan **map** dengan **lambda**
2. Memfilter produk yang stoknya > 0 (tersedia) menggunakan **filter** dengan **lambda**
3. Memfilter produk berdasarkan kategori tertentu menggunakan **filter** dengan **lambda**
4. Mengekstrak list nama produk yang harganya di atas 1 juta menggunakan kombinasi **filter** dan **map**

### C. Closure - Factory Function (25 poin)
Buatlah closure/factory function:
1. `create_diskon_calculator(persen)` - Mengembalikan fungsi yang menghitung harga setelah diskon
   - Contoh: `diskon_10 = create_diskon_calculator(10)` akan memberikan diskon 10%
   - `diskon_10(1000000)` menghasilkan `900000`

2. `create_kategori_filter(kategori)` - Mengembalikan fungsi yang memfilter produk berdasarkan kategori
   - Contoh: `filter_elektronik = create_kategori_filter("Elektronik")`

3. `create_harga_range_filter(min_harga, max_harga)` - Mengembalikan fungsi yang memfilter produk dalam range harga tertentu
   - Contoh: `filter_harga = create_harga_range_filter(500000, 2000000)`

### D. Sequence Type Operations (15 poin)
Gunakan **list comprehension** dan **dictionary comprehension** untuk:
1. Membuat list tuple `(nama, harga, stok)` untuk semua produk
2. Membuat dictionary `{id: nama}` untuk semua produk
3. Membuat dictionary `{kategori: [list_produk]}` yang mengelompokkan produk berdasarkan kategori

### E. Reduce untuk Agregasi & Pipeline (15 poin)
Gunakan **reduce** untuk:
1. Menghitung total nilai inventory (sum dari harga x stok semua produk)
2. Menghitung total nilai transaksi (berdasarkan data transaksi dan produk)
3. Mencari produk dengan nilai stok tertinggi
