**c.** Berikut adalah kode dengan konsep Pure Function :

```python
def double_all_recursive(numbers):
    if len(numbers) == 0:
        return []
    return [numbers[0] * 2] + double_all_recursive(numbers[1:])

numbers = [2, 4, 6, 8, 10]
double_numbers = double_all_recursive(numbers)
print(double_numbers)
```

### Jawaban Soal 2:

**a.** Dari hasil eksekusi program di atas, output yang saya dapatkan adalah `45`. Hal tersebut terjadi karena kode tersebut melakukan penjumlahan dari semua elemen yang ada dalam `my_list = [3, 6, 9, 12, 15]` yaitu 3 + 6 + 9 + 12 + 15 sehingga menghasilkan hasil akhir nilai 45.

**b.** Berikut adalah fungsi yang saya buat untuk memberikan output factorial untuk setiap nilai dari my_list:



```python
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)

my_list = [3, 6, 9, 12, 15]
factorials = [factorial_recursive(x) for x in my_list]

print("List asli:", my_list)
print("Hasil faktorial:", factorials)

for num, fact in zip(my_list, factorials):
    print(f"Faktorial dari {num} adalah: {fact}")
```

Program tersebut menghitung faktorial setiap elemen dalam list menggunakan pendekatan rekursif. Fungsi `factorial_recursive(n)` menggunakan base case (0 atau 1 menghasilkan 1) dan recursive case. List comprehension menerapkan fungsi ini pada setiap elemen sehingga menghasilkan list faktorial. Hasil kemudian ditampilkan dengan menampilkan list asli, list hasil faktorial, dan detail setiap perhitungan menggunakan loop `zip()`.

**c.** Berikut adalah modifikasi kode penjumlahan menggunakan gaya penulisan fungsional dengan rekursif yang saya buat:



```python
def sum_recursive(lst):
    if len(lst) == 0:
        return 0
    return lst[0] + sum_recursive(lst[1:])

my_list = [3, 6, 9, 12, 15]
result = sum_recursive(my_list)
print(f"Hasil penjumlahan: {result}")
```

### 3. Perhatikan program di bawah ini !

```python
def fun(x):
    for i in range(1,x) :
        yield x*i
y = fun(5)

print(y)
```

### Jawaban Soal 3:

**a.** Berdasarkan analisis saya terhadap program di atas, program tersebut menggunakan konsep Deklaratif (Fungsional) karena memakai yield yang merupakan fitur pemrograman fungsional. Generator dengan yield mengadopsi lazy evaluation dan tidak secara eksplisit memutasi state. Kode fokus pada "apa yang dihasilkan" (urutan nilai) alih-alih "bagaimana cara menghitungnya" secara langkah-demi-langkah. Meski ada for loop di dalam fungsi, pola generator dengan yield secara keseluruhan bersifat deklaratif/fungsional.

**b.** Ketika saya menjalankan perintah `print(y)`, output yang saya dapatkan adalah `<generator object fun at 0x0000020F3BCA1630>`. Output ini menunjukkan bahwa `y` adalah sebuah generator object, bukan nilai atau list hasil perhitungan. Generator object ini berisi instruksi untuk menghasilkan nilai secara bertahap, namun nilai-nilai tersebut belum dihitung sampai generator di-iterate. Oleh karena itu, saat kita mencetak generator secara langsung, Python hanya menampilkan representasi objeknya (nama fungsi dan alamat memori), bukan nilai yang dihasilkannya.

**c.** Generator bersifat lazy evaluation. Saat fun(5) dipanggil, fungsi tidak langsung menghitung dan menyimpan semua nilai; fungsi hanya mengembalikan generator object yang siap untuk di-iterate. yield membuat fungsi menjadi generator yang memproduksi nilai on-demand saat di-iterate, bukan menyimpan semua nilai sekaligus di memori. Ketika print(y) dipanggil, Python menampilkan representasi string dari generator object, bukan nilai yang dihasilkannya. Untuk mendapatkan nilai-nilai tersebut, generator perlu di-iterate menggunakan next(), loop for, atau dikonversi ke list(y). Dengan demikian, output <generator object fun at ...> muncul karena generator adalah object iterator yang belum dieksekusi secara penuh.
