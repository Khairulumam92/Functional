### Codelab 1

Lengkapi program dibawah ini untuk memisahkan data harga yang diinput oleh user menjadi dua kategori:

* Harga murah: kurang dari 50.000, disimpan ke dalam list

* Harga mahal: sama dengan atau lebih dari 50.000 dan disimpan ke dalam tuple

 Input dipisahkan dengan spasi, dan kamu harus melakukan casting sesuai tipe data yang dibutuhkan.


```python
# Moh. Khairul Umam - 448 

def pisahkan_harga(input_harga):
    list_murah = []
    list_mahal = ()

    # Proses setiap harga
    for x in input_harga.split():
        harga = int(x)  # <-- TODO 1: casting ke integer
        if harga < 50000:  # <-- TODO 2: kondisi harga murah
            list_murah.append(harga)  # <-- TODO 3: tambahkan ke list
        else:
            list_mahal += (harga,)  # <-- TODO 4: tambahkan ke tuple
    return list_murah, list_mahal


input_harga = input("Masukkan daftar harga (pisahkan dengan spasi): ")

murah, mahal = pisahkan_harga(input_harga)

print("Harga murah:", murah)
print("Harga mahal:", mahal)
```

### Codelab 2
Sebuah program menghitung total belanja berdasarkan data barang dan harga.

1. Buatlah sebuah dictionary yang menyimpan data nama barang dan harganya.

2. Lengkapi fungsi untuk menghitung total belanja berdasarkan isi dictionary.

3. Tampilkan total belanja.

4. Lengkapi dan perbaiki kode berikut agar dapat berjalan dengan benar.



```python
# Moh. Khairul Umam - 448

# TO DO 1: Lengkapi dictionary berikut dengan nama barang dan harganya (minimal 3 data)
barang = {
    "Casing": 400000,
    "GPU 9080": 15000000,
    "Monitor Lenovo": 10000000
}

# TO DO 2: Lengkapi dan betulkan func hitung_total belanja berikut
def hitung_total(data):
    total = 0
    for item in data:  
        harga = data[item]   # ambil harga berdasarkan key
        total += harga
    return total

# TO DO 3: tampilkan hasil total belanja
hasil = hitung_total(barang)
print("Total belanja Anda adalah Rp:",hasil)
```
